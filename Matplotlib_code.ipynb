```python
#ploting in pandas
normals = pd.Series(np.random.normal(size=50))
print(normals)
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
variables = pd.DataFrame({'normal':np.random.normal(size=20),
                         'gamma':np.random.gamma(1,size=20),
                         'poisson':np.random.poisson(size=20)})
print(variables)
variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(subplots=True)
```

```python
variables.cumsum(0).plot(secondary_y='normal')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1,figsize=(4,14))
for i, var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i],title=var)
    axes[0].set_ylabel('cumulative sum')
```

```python
#simple line plots
fig = plt.figure()
ax = plt.axes()
s = np.linspace(0,10,50)
print(s)
ax.plot(s,np.sin(s));
```

```python
plt.plot(s,np.sin(s))
```

```python
#line colors and styles
plt.plot(s,np.sin(s-3),color='red')
plt.plot(s,np.sin(s-1),color='y')
plt.plot(s,np.sin(s-2),color='0.66')
```

```python
plt.plot(s,np.sin(s-0),linestyle='-')
plt.plot(s,np.sin(s-1),linestyle=':')
plt.plot(s,np.sin(s-2),linestyle='-.')
```

```python
plt.plot(s,np.sin(s-0))
plt.xlim(-4,15)
plt.ylim(-3,2);
```

```python
#labelling plots
plt.plot(s,np.cos(s))
plt.title("Cosin Curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

```python
rng = np.random.RandomState(7)
x = rng.randn(200)
y = rng.randn(200)
colors = rng.rand(200)
sizes = 1000*rng.rand(200)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,
           cmap='Reds')
plt.colorbar();
```

```python
x = np.linspace(0,10,200)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(200)
plt.errorbar(x,y,yerr=dy,fmt='.b');
```

```python
#2 dimensional histogram
mean =[5,2]
cov = [[3,1],[2,0,5]]
x,y = np.random.multivariate_normal(mean,cov,10000)
```

```python
plt.hist2d(x, y, bins=40, cmap='Reds')
cb = plt.colorbar()
```
